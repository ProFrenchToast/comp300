```python
from comp300.LearningModel.AgentClasses import *
from comp300.Chess.ChessWrapper import *
import gym
from baselines.common.cmd_util import make_vec_env
import torch
from baselines.run import *
gym.register(id='ChessSelf-v0',
             entry_point='comp300.Chess.ChessWrapper:ChessEnv',
             max_episode_steps=1000)
```

```python
#calculate the winrate for an agent
def getResult(env, agent):
    r = testRun(env, agent)
            
    if r == 100 or r == 0:
        return 1
    else:
        return 0
    return  totalReward

def testRun(env, model):
    obs = env.reset()
    state = model.initial_state if hasattr(model, 'initial_state') else None
    dones = np.zeros((1,))

    episode_rew = np.zeros(env.num_envs) if isinstance(env, VecEnv) else np.zeros(1)
    while True:
        if state is not None:
            actions, _, state, _ = model.step(obs,S=state, M=dones)
        else:
            actions, _, _, _ = model.step(obs)

        obs, rew, done, _ = env.step(actions)
        episode_rew += rew
        #env.render()
        done_any = done.any() if isinstance(done, np.ndarray) else done
        if done_any:
            return rew[0]
                
    return 0


args = []
args.append("--alg=ppo2")
args.append("--env=ChessSelf-v0")
args.append("--num_timesteps=0")
args.append("--load_path=/home/patrick/models/chessTest/chess20Mppo2")
args.append("--network=mlp")
args.append("--num_env=1")
#args.append("--num_layers=5")

env_type = "ChessWrapper"

arg_parser = common_arg_parser()
args, unknown_args = arg_parser.parse_known_args(args)
extra_args = parse_cmdline_kwargs(unknown_args)

model, env = train(args, extra_args)

agent = model

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
network = RewardNetwork("", env=env, env_type=env_type)
network.to(device)
```

```python
numEpisodes = 1000

RLwins = 0
RLresults = []
for i in range(numEpisodes):
    reward = testRun(env, agent)
    RLwins +=reward
    print(reward)
    
print(RLwins)
```
